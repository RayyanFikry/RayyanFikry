## Mapping

```python
time_map_1_4 = {
    '0 - 2 Jam': 1,
    '2 - 4 Jam': 2,
    '4 - 6 Jam': 3,
    '> 6 Jam': 4
}

sleep_map = {
    '2 - 4 Jam': 1,
    '4 - 6 Jam': 2,
    '6 - 8 Jam': 3,
    '> 8 Jam': 4
}

work_map = {
    'Tidak': 0,
    'Iya': 1
}

df['Jam Belajar UTS/UAS (1-4)'] = df['Jam Belajar UTS/UAS (1-4)'].map(time_map_1_4)
df['Sosial Media (1-4)'] = df['Sosial Media (1-4)'].map(time_map_1_4)
df['Ngatur Waktu Kuliah, Tugas, Belajar (1-5)'] = df['Ngatur Waktu Kuliah, Tugas, Belajar (1-5)'].astype(int)
df['Mudah memahami materi (1-5)'] = df['Mudah memahami materi (1-5)'].astype(int)
df['Notes/ringkasan Kuliah (1-5)'] = df['Notes/ringkasan Kuliah (1-5)'].astype(int)
df['Jadwal belajar teratur (1-5)'] = df['Jadwal belajar teratur (1-5)'].astype(int)
df['Memanfaatkan AI (1-5)'] = df['Memanfaatkan AI (1-5)'].astype(int)
df['Revisi Pelajaran (1-5)'] = df['Revisi Pelajaran (1-5)'].astype(int)

df['Jumlah Organisasi (0-3)'] = (
    df['Jumlah Organisasi (0-3)']
      .replace({'> 3': 3})
      .astype(int)
)

df['Terbebani Organisasi (1-5)'] = df['Terbebani Organisasi (1-5)'].astype(int)

df['Kerja Samping (0/1)'] = df['Kerja Samping (0/1)'].map(work_map)

df['Kerja Sampingan mengganggu (1-5)'] = df['Kerja Sampingan mengganggu (1-5)'].astype(int)
df['Medsos Mengganggu (1-5)']         = df['Medsos Mengganggu (1-5)'].astype(int)

df['Waktu Tidur (1-4)'] = df['Waktu Tidur (1-4)'].map(sleep_map)
df['Waktu Tidur Cukup (1-5)'] = df['Waktu Tidur Cukup (1-5)'].astype(int)
```
